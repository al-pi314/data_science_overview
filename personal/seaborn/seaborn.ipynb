```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], bins=20, kde=False)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

KDE = Kernel density estimation -> line on distplot (without kde = False)

Construction:   
- create gaussian distribution plot on each of the points of
- sum all gaussian distributions plots
    
It is analogous to histograms.

```python
sns.kdeplot(tips['tip'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, jitter=True, dodge=True)
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, dodge=True, color='black')
```
