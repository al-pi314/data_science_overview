```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips_corr = tips.corr()
tips_corr
```

```python
sns.heatmap(tips_corr, annot=True, cmap='coolwarm')
```

```python
flight_passegers = flights.pivot_table(index='month', columns='year', values='passengers')
flight_passegers
```

```python
sns.heatmap(flight_passegers, cmap='magma', linecolor='white', linewidths=0.05)
```

```python
sns.clustermap(flight_passegers, cmap='coolwarm', standard_scale=1)
```
