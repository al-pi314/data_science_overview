```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x, y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object oriented
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left offset, bottom offset, width, height (all in procentage notation)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.3, 0.3])
axes2 = fig.add_axes([0.4, 0.4, 0.9, 0.9])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title("LARGER PLOT")

axes2.plot(y, x)
axes2.set_title("SMALLER PLOT")
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
```

```python
axes # all axes on a figure (from left to right, top to bottom)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
axes[0, 0].plot(x, y)
axes[1, 1].plot(y, x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 2), dpi=500)

axes[0, 1].plot(x, y)

axes[1, 0].plot(y, x)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(3,3), dpi=200)
axes.plot(x, x**2, label="X Sqaured")
axes.plot(x, x**0.5, label="Root of X")

axes.legend(loc=(0.1, 0.7)) # if location is not set the method tries to minimize overlaying
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 0.45, 1])
ax2 = fig.add_axes([0.5, 0, 0.45, 1])

# plot styles
ax1.plot(x, y, color="#f022cc", linewidth=5, alpha=0.5) # hex code or common names
ax2.plot(   x, y, 
            lw=0.1, color="purple", ls="dashed", 
            marker='o', ms=10, markerfacecolor="yellow", markeredgecolor="red", markeredgewidth=2
        )

ax1.set_xlim([3, 5])
ax1.set_ylim([10, 20])

```
